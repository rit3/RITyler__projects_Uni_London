## Introduction
### Overview of topic
As investment in artificial intelligence (AI) increases, its important to establish public perception. Trust in technology is the cornerstone of acceptance of AI. Despite its benefits to society, there are concerns that have been identified, these include: regulation of AI, data abuse, job loss and AI system dependence.  The media has a large role to play in public perception of AI, it may influence fear about the technology, particularly in those who are digitally unaware.
It is therefore import to understand what public perception about AI is.  This survey attempts to discover what people really think about AI.

### Research objectives:
The study aims to conduct a comparative analysis of the public's attitude towards AI based on data from an online survey and then compare it to a study made by KPMG and the University of Queensland, Australia. 

The population chosen for this study is the general public, mostly from South Africa, but includes anyone who was willing to answer the survey questionnaire. 

The survey was shared on various social media groups on Facebook, LinkedIn, Whatsapp, the University of London's Slack channel as well as on the University of London's Data Visualization student course area.

### Motivation for the study:
The public attitude toward AI is a very important aspect of AI technology and its acceptance of it. The motivation for this study is to collect information about how the public feels about AI and their level of trust in it. The survey will also study the governance of AI and its perceived risks. It will attempt to determine if demographics influence AIs' acceptance. Trust is at the heart of public acceptance of AI. Trustworthy AI is the most important factor in many studies about public perceptions of AI. Data privacy, security, governance, and accountability are some of the issues that many expect from organisations administering AI. 

### Overview of key findings
- Most people in the survey have a basic understanding of AI (63%).
(More university educated people had a higher understanding as did males in the age group 18 - 34; those in the IT, self-employed and student sectors had a higher than average understanding of AI; No one in the survey claimed to have expert knowledge in AI)
- Everyone in the survey used some form of AI technology. 
(Social media, online banking, navigation, google search, and smartphone are the top 5 most commonly used apps)
- 86% of survey respondents are positive about AI in general
(Internet of Things, 3D printing, cloud computing, VR and AR were also seen in a positive light)
- No one rejected AI at all.
(78% approved, accepted or embraced AI)
- No one is completely unwilling to share information with AI.
- Almost half of all respondents believe the benefits of AI outweigh risks.
- Over half of all respondents use AI in the workplace.
- Respondents have a high level of comfort with automated admin processes, assistance with queries, automating data pattern detection and interpretation, automating physical tasks and marketing activities as well as monitoring digital and physical security.
- Most people think there are some problems with AI.
(Digital manipulation, cyber-attacks, autonomous weapons, data privacy and hiring bias were the top 5 problems associated with AI)
- 75% have confidence in universities to manage the development and use of AI in the best interests of the public.**

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
